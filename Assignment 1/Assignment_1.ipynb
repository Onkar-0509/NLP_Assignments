# Import Required Libraries

# Sample Dataset

```python
text = "I don't like natural language processing! Loving NLP üòç #AI @OpenAI"
```

 # Whitespace Tokenizer

```python
# Whitespace Tokenizer
from nltk.tokenize import WhitespaceTokenizer

tokenizer = WhitespaceTokenizer()
tokens = tokenizer.tokenize(text)

print(tokens)
```

# Punctuation Tokenizer

```python
# Punctuation Tokenizer
from nltk.tokenize import wordpunct_tokenize

tokens = wordpunct_tokenize(text)
print(tokens)
```

# Treebank Tokenizer

```python
# Treebank Tokenizer
from nltk.tokenize import TreebankWordTokenizer

tokenizer = TreebankWordTokenizer()
tokens = tokenizer.tokenize(text)

print(tokens)


```

# MWE (Multi-Word Expression) Tokenizer

```python
# MWE (Multi-Word Expression) Tokenizer
from nltk.tokenize import MWETokenizer

tokenizer = MWETokenizer([('natural', 'language', 'processing')], separator='_')
tokens = tokenizer.tokenize(text.split())

print(tokens)
```

# Porter Stemmer

```python
# Porter Stemmer
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

words = ['running', 'connected', 'studies']
stemmed_words = [stemmer.stem(word) for word in words]

print(stemmed_words)
```

# Snowball Stemmer

```python
#Snowball Stemmer
from nltk.stem import SnowballStemmer

stemmer = SnowballStemmer("english")

words = ['running', 'connected', 'studies']
stemmed_words = [stemmer.stem(word) for word in words]

print(stemmed_words)
```

# WordNet Lemmatizer

```python
# WordNet Lemmatizer
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

words = ['running', 'better', 'studies']
lemmatized_words = [lemmatizer.lemmatize(word) for word in words]

print(lemmatized_words)
```

# (Optional ‚Äì Better Lemmatization with POS)

```python
#(Optional ‚Äì Better Lemmatization with POS)
lemmatized_words = [
    lemmatizer.lemmatize('running', pos='v'),
    lemmatizer.lemmatize('better', pos='a'),
    lemmatizer.lemmatize('studies', pos='n')
]

print(lemmatized_words)
```
