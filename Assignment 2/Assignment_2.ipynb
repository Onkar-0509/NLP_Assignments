# Import Required Libraries

```python
import numpy as np
import pandas as pd

from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from gensim.models import Word2Vec
```

# Sample Dataset

```python
documents = [
    "I love natural language processing",
    "Natural language processing is fun",
    "I love NLP"
]

print(documents)
```

# Bag of Words – Count Occurrence

```python
count_vectorizer = CountVectorizer()
bow_counts = count_vectorizer.fit_transform(documents)

bow_df = pd.DataFrame(
    bow_counts.toarray(),
    columns=count_vectorizer.get_feature_names_out()
)

bow_df
```

# Bag of Words – Normalized Count Occurrence

```python
from sklearn.feature_extraction.text import CountVectorizer

count_vectorizer = CountVectorizer()
bow_counts = count_vectorizer.fit_transform(documents)
```

```python
from sklearn.preprocessing import normalize
import pandas as pd

bow_normalized = normalize(bow_counts, norm='l1')

bow_norm_df = pd.DataFrame(
    bow_normalized.toarray(),
    columns=count_vectorizer.get_feature_names_out()
)

bow_norm_df
```

# TF-IDF (Term Frequency – Inverse Document Frequency)

```python
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(documents)

tfidf_df = pd.DataFrame(
    tfidf_matrix.toarray(),
    columns=tfidf_vectorizer.get_feature_names_out()
)

tfidf_df
```

# Tokenize Text for Word2Vec

```python
tokenized_docs = [doc.lower().split() for doc in documents]
tokenized_docs
```

# Word Embedding for a Word

```python
word2vec_model = Word2Vec(
    sentences=tokenized_docs,
    vector_size=50,
    window=3,
    min_count=1,
    workers=4
)
```

```python
word2vec_model.wv['language']
```

# Similar Words using Word2Vec

```python
word2vec_model = Word2Vec(
    sentences=tokenized_docs,
    vector_size=50,
    window=3,
    min_count=1,
    workers=4
)
```

```python
word2vec_model.wv['language']
```

```python
word2vec_model.wv.most_similar('language')
```

# Sentence Embedding (Average Word Vectors)

```python
def sentence_embedding(sentence, model):
    words = sentence.lower().split()
    vectors = [model.wv[word] for word in words if word in model.wv]
    return np.mean(vectors, axis=0)

sentence_vec = sentence_embedding("I love natural language processing", word2vec_model)
sentence_vec
```
